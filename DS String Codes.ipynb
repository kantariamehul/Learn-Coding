```python
# Pangram Checking
def checkPangram(s):
  if len(s)<26:
    return 0
  Flag=[0]*26
  for c in s:
    if(c>='A' and c<='Z'):
      c=c.lower()
    if(c>='a' and c<='z'):
      loc=ord(c)-ord('a')
    Flag[loc]=1
  if(sum(Flag)==26):
    return 1
  else:
    return 0

St="The quick brown fox jumps over the little lazy dog"
print(checkPangram(St))
```
