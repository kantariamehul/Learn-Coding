```python
# Rotate
Ar=[x for x in range(10)]
print("Original:",Ar)

def ArrayRotate1(Ar,r): # Space: O(n), Time: O(n)
  Temp=Ar[r::]
  Temp.extend(Ar[0:r])
  return Temp

print("M1:",ArrayRotate1(Ar,2))
print("M1:",ArrayRotate1(Ar,-2))

def ArrayRotate2(Ar,r): # Space: O(1), Time: O(n*r)
  if(r>0):
    while (r>0):
      First=Ar[0]
      for i in range(0,len(Ar)-1):
        Ar[i]=Ar[i+1]
      Ar[i+1]=First
      r-=1
  else:
    while (r<0):
      Last=Ar[-1]
      for i in range(0,len(Ar)-1):
        Ar[len(Ar)-i-1]=Ar[len(Ar)-i-2]
      Ar[0]=Last
      r+=1
  return(Ar)

Ar=[x for x in range(10)]
print("M2:",ArrayRotate2(Ar,2))

Ar=[x for x in range(10)]
print("M2:",ArrayRotate2(Ar,-2))

def ArrayRotate3(Ar,d): #Space: O(1), Time: O(n)
  ReverseArray(Ar,0,d-1)
  ReverseArray(Ar,d,len(Ar)-1)
  ReverseArray(Ar,0,len(Ar)-1)
  return Ar

def ReverseArray(Ar,S,E):
  while(S<E):
    Ar[S],Ar[E]=Ar[E],Ar[S]
    S+=1
    E-=1

Ar=[x for x in range(10)]
print("M3:",ArrayRotate3(Ar,2))
```
