```python
#read/open a file without with
#Read mode, Can not edit
File1_r=open("Hello.txt","r")
File1_r.close()

#Append mode
File1_a=open("Hello.txt","a")
File1_a.write("Opened in append mode")
File1_a.close()

# write mode, clear all content and start writing
File1_w=open("Hello.txt","w")
File1_w.write("Writen in Write mode")
File1_w.close()

# with try and finally
File1_w=open("Hello.txt","w")
try:
    File1_w.write("With try and finally")
finally:
    File1_w.close()
```

```python
#file open using with
with open("Hello.txt","w") as File2_w:
    File2_w.write("Uisng with")
# no need to close the file - Difference
```

```python
#class
class Car:
    Speed=0 # Attributes, Data 
    def Acc(self): # Method
        self.Speed+=5
    def Break(self): # Method
        self.Speed-=5
SUV=Car()
print(SUV.Speed)
SUV.Acc()
print(SUV.Speed)
SUV.Acc()
print(SUV.Speed)
SUV.Break()
print(SUV.Speed)
```
