```python
#Heap = Max or Min heap (its a complete binary search tree)
# A[i] array is used to design heap
# A[i] is parent of left node A[2*i+1] and right node A[2*i+2]
# A[int((i-1)/2)] is parent of A[i]


from heapq import heappush, heappop, heapify

class MinHeap:
  def __init__(self):
    self.heap=[]
  
  def parent(self,i):
    return (i-1)//2

  def insertkey(self,k):
    heappush(self.heap,k)
  
  def decreasekey(self,i,new_val):
    self.heap[i]=new_val
    while(i != 0 and self.heap[self.parent(i)] > self.heap[i]):
      self.heap[i],self.heap[self.parent(i)]=self.heap[self.parent(i)],self.heap[i]

  def extractMin(self):
    return heappop(self.heap)

  def getMin(self):
    return self.heap[0]
  
  def print(self):
    print(list(x for x in self.heap))


mHeap=MinHeap()
mHeap.insertkey(15)
mHeap.insertkey(20)
mHeap.insertkey(3)
mHeap.insertkey(45)
mHeap.insertkey(5)
mHeap.insertkey(1)
mHeap.insertkey(4)
  
print("Heap is:")
mHeap.print()

mHeap.decreasekey(4,2)
mHeap.print()

mHeap.extractMin()
mHeap.print()

mHeap.getMin()
mHeap.print()


```

```python
# Graph with Adjuncy list
class AdjNode:
  def __init__(self,data):
    self.node=data
    self.next=None

class Graph:
  def __init__(self,NoofNodes):
    self.NoofNodes=NoofNodes
    self.graph = [None]* self.NoofNodes

  def add_edge(self,src,dest):
    node = AdjNode(dest)
    node.next=self.graph[src]
    self.graph[src] = node

    node = AdjNode(src)
    node.next=self.graph[dest]
    self.graph[dest] = node

  def print_graph(self):
    for i in range(self.NoofNodes):
      print(i,'->',end="")
      temp=self.graph[i]
      while (temp):
        print(temp.node,'->',end="")
        temp=temp.next
      print("NULL\n")

G=Graph(5)
G.add_edge(0,1)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)

G.print_graph()


```

```python
# Double Link List
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
    self.prev=None

class LinkedList:
  def __init__(self):
    self.head=None

  def PrintLL(self):
    print("The LL is:",end=" ")
    Temp=self.head
    Last=Temp.prev
    while(Temp!=Last):
      print(Temp.data,end=" ")
      Temp=Temp.next
    print(Temp.data)

  def PrintLLReverse(self):
    Temp=self.head
    First=Temp.next
    while(Temp!=First):
      print(Temp.data,end=" ")
      Temp=Temp.prev
    print(Temp.data)

  
  def Push(self,data):
    if (self.head==None): # add as head node
      Head=Node(data)
      self.head=Head
      Head.next=Head
      Head.prev=Head
    else:
      Last=self.head
      Head=self.head
      while (Last.next!=Head):
        Last=Last.next
      Temp=Node(data)
      Last.next=Temp
      Temp.next=Head
      Head.prev=Temp
      Temp.prev=Last
    self.PrintLL()

  def Search(self,data):
    Loc=1
    Temp=self.head
    Last=self.head.prev
    while(Temp.data!=data):
      Loc+=1
      Temp=Temp.next
    print("Location of data",data,"is Node",Loc)

  def DeleteNode(self,data):
    Temp=self.head
    while(Temp.data!=data):
      Temp=Temp.next
    Temp.prev.next=Temp.next
    Temp.next.prev=Temp.prev

  def PrintNodeDetails(self):
    Temp=self.head
    Last=Temp.prev
    while(Temp!=Last):
      print("Data:",Temp.data,"Prev:",Temp.prev,"Next:",Temp.next)
      Temp=Temp.next

LL=LinkedList()
LL.Push(0)
LL.Push(1)
LL.Push(2)
LL.Push(3)
LL.Push(4)
LL.Search(3)
LL.PrintNodeDetails()
LL.PrintLLReverse()
LL.DeleteNode(3)
LL.PrintLL()
```
